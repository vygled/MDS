### Задание 1


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.


```python
import numpy as np
from sympy import cos, diff, sin, Symbol
```

```python
x = Symbol('x')
f = x**5 + 4*sin(2*x) + cos(3*x+3)

first_derivative = diff(f)
second_derivative = diff(first_derivative)

x1 = 1
fd1 = first_derivative.evalf(subs={x:x1})
sd1 = second_derivative.evalf(subs={x:x1})


print(f'В точке x = {x1}:')
print(f'значение первой производной: {round(fd1, 2)},')
print(f'значение второй производной: {round(sd1, 2)}.')
```

### Задание 2

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
import matplotlib.pyplot as plt
```

```python
x = Symbol('x')

f = (sin(2*x+1))**5
first_derivative = diff(f)
second_derivative = diff(first_derivative)

display(first_derivative)
display(second_derivative)

vx  = np.arange(-5, 5, 0.01)
vy1 = []
vy2 = []

for n in vx:
    vy1.append(first_derivative.subs(x, n))
    vy2.append(second_derivative.subs(x, n))

plt.plot(vx, vy1, label='Первая производная')
plt.plot(vx, vy2, label='Вторая производная')
plt.legend()
plt.show() 
    
```
