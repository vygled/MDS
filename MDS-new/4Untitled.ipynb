```python
def function(x):
    return 1/(1-x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,2*(np.pi), 0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.plot(x, function(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('FUNCTION')
plt.show()
```
